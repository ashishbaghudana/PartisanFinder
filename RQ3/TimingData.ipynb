#Build Website Lists

For each political leaning page on MediaBiasFactCheck, scrape the page for URLs (which link to other MBFC pages), request those pages, and get the actual website URLs from those pages.

RUN THIS FIRST
The other cells depend on these MBFC lists

```python
import requests
from bs4 import BeautifulSoup

def load_site(u):
  """Retrieves all URLs from a political leaning page on MediaBiasFactCheck"""
  sitelist = []
  url = u
  r = requests.get(url)
  soup = BeautifulSoup(r.text, "lxml")
  sources = soup.find('div', {'class': 'entry clearfix'}).find_all('a', {'class': ''})

  for source in sources:
    name = source.text
    _url = source['href']
    
    if _url == 'https://mediabiasfactcheck.com/euromaiden-press/':
      continue
    if _url == 'http://mediabiasfactcheck.com/afp-agence-france-presse/':
      continue
    if _url == 'https://mediabiasfactcheck.com/portland-press-herald/':
      continue
    if _url == 'https://mediabiasfactcheck.com/relevant-magazine/':
      continue
    if _url == 'https://www.hermancain.com/':
      continue
    if _url == 'https://mediabiasfactcheck.com/western-free-press/':
      continue
    
    try:
        _r = requests.get(_url)
        _soup = BeautifulSoup(_r.text, "lxml")
    
        for p in _soup.find('div', {'class': 'entry-content'}).find_all('p'):
          if p.text.startswith('Source'):
        
            for a in p.find_all('a'):
              __url = a['href']
              #sitelist.append(p.text.split('//')[-1].strip())
              #print(p.text.split('//')[-1].strip())
              sitelist.append(__url)
              print(__url)
    
    except:
        print('Error retrieving from {0}'.format(_url))
        
  print('{0} websites found'.format(len(sitelist)))
  return sitelist

print('Processing left-leaning sites...')
websites_left = load_site('https://mediabiasfactcheck.com/left/')
print('Processing center-left-leaning sites...')
websites_leftcenter = load_site('https://mediabiasfactcheck.com/leftcenter/')
print('Processing center-leaning sites...')
websites_center = load_site('https://mediabiasfactcheck.com/center/')
print('Processing center-right-leaning sites...')
websites_rightcenter = load_site('https://mediabiasfactcheck.com/right-center/')
print('Processing right-leaning sites...')
websites_right = load_site('https://mediabiasfactcheck.com/right/')
print('Done')
```

Start recording load times

For each of the websites that we found on MBFC, run 10 trials to see how long the page takes to load. Save in a table.

```python
from collections import namedtuple
import pandas as pd
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import signal

### Start timing trials for a bias category
def processCategory(c, a):
  """Run load time trials on an entire political leaning category of websites"""
  for website in c:
    processWebsite(website, a)
#   for k in range(250,384):
#     processWebsite(c[k],a)


### Start timing trials for a single website
def processWebsite(u, a):
  """Run 10 load time trials on the current website"""
  times = []
  for i in range(0, 10):
    try: 
#       response = requests.get(u, timeout=2.5)
#       times.append(response.elapsed.total_seconds())
      times.append(singleTrialSelenium(u))
      print('{0}: Trial #{1} -- {2}'.format(u, i, times[i]))
    except:
      print('error on {0}, trial #{1}'.format(u,i))
      times.append(9999.99)
    
  timingData.append(DataContainer(url=u, align=a, t0=times[0], t1=times[1], t2=times[2], t3=times[3], t4=times[4],
                                  t5=times[5], t6=times[6], t7=times[7], t8=times[8], t9=times[9]))
    
### Run a single timing trial    
def singleTrialSelenium(u):
  driver = webdriver.PhantomJS()
  driver.set_window_size(1120, 550) 
  #driver.manage().timeout().pageLoadTimeout(200, TimeUnit.SECONDS)
  driver.implicitly_wait(200)
  driver.set_page_load_timeout(200)
  startTime = time.time()
  driver.get(u);
  #element = WebDriverWait(dr, 10).until(EC.presenceOfElementLocated(By.id("Calculate")));
  endTime = time.time()
  driver.service.process.send_signal(signal.SIGTERM)
  return endTime - startTime
    
DataContainer = namedtuple('DataContainer', ['url', 'align', 't0', 't1', 't2', 't3', 't4', 't5', 't6', 't7', 't8', 't9'])
timingData = []

  
processCategory(websites_left, "L")
processCategory(websites_leftcenter, "LC")
processCategory(websites_center, "C")
processCategory(websites_rightcenter, "RC")
processCategory(websites_right, "R")

df = pd.DataFrame(timingData, columns=DataContainer._fields)
file_path = 'TimingData.csv'
df.to_csv(file_path, sep=',', encoding='utf-8')
print('CSV written')
```

Iterate through the MBFC bias lists, trying to scrape a US pagerank score for each of the websites

```python
"""Get Alexa scores for each of the websites"""

from collections import namedtuple
import pandas as pd
import requests
from bs4 import BeautifulSoup

def processCategory(c, a):
  """Obtain Alexa scores for an entire political leaning category of websites"""
  for website in c:
    processWebsite(website, a)

def processWebsite(w, a):
  u = createAlexaURL(minimizeURL(w))  

  try: 
    response = requests.get(u, timeout=2.5)
    s = parseAndRetrieve(response)
    print('{0}: Score is {1}'.format(u, s))
  except:
    print('error retrieving score for {0}'.format(u))
    s = 99999999
    
  alexaData.append(DataContainer(url=w, align=a, score=s))  

def minimizeURL(w):
  """Turn 'http://www.cnn.com' into 'cnn.com' """
  beg = w.split('/')[2].strip()
  print(beg)  
  pt1 = beg.split('.')[-2].strip()
  pt2 = beg.split('.')[-1].strip()  
  return pt1+'.'+pt2

def createAlexaURL(w):
  """Turn 'cnn.com' into 'https://www.alexa.com/siteinfo/cnn.com' """
  alexa = 'https://www.alexa.com/siteinfo/'
  return alexa+w

def parseAndRetrieve(r):
  """parse the website using beautiful soup; look for the alexa score"""
  soup = BeautifulSoup(r.text, "lxml")
  for p in soup.find('span', {'class': 'countryRank'}).find_all('h4', {'class': 'metrics-title'}):
    if 'United States' not in p.text.strip():
      print(p.text.strip())  
      return 99999999
  for p in soup.find('span', {'class': 'countryRank'}).find_all('strong', {'class': 'metrics-data'}):
    return p.text.strip()


    
DataContainer = namedtuple('DataContainer', ['url', 'align', 'score'])
alexaData = []
  
processCategory(websites_left, "L")
processCategory(websites_leftcenter, "LC")
processCategory(websites_center, "C")
processCategory(websites_rightcenter, "RC")
processCategory(websites_right, "R")

# processWebsite('http://www.abs-cbn.com/','L')
# processWebsite('http://www.cnn.com/something/somethingelse/','L')

df = pd.DataFrame(alexaData, columns=DataContainer._fields)
file_path = 'AlexaScores.csv'
df.to_csv(file_path, sep=',', encoding='utf-8')
print('CSV written')
```
