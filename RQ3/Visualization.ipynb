```python
import seaborn as sns
import numpy as np
import csv
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
data = []
with open('../Visualizations/violinplot.csv') as freader:
    csvreader = csv.reader(freader)
    for row in csvreader:
        data.append(np.array(row, dtype=np.float32))
data = np.array(data[:-1])
print(data.shape)
```

```python
sns.set_style('whitegrid')
fig, ax = plt.subplots(figsize=(8, 5), dpi=150)
sns.set(font_scale = 1)
ax = sns.violinplot(data=data, orient='h', ax=ax)
ax.set(yticklabels=['L\n(N=73)', 'LC\n(N=172)', 'C\n(N=96)', 'RC\n(N=73)', 'R\n(N=57)'])
ax.set_xlabel('Load time (in seconds)', fontsize=10)
ax.set_ylabel('Category', fontsize=10)
fig = ax.get_figure()
fig.savefig('../Visualizations/violinplot.png')
```
